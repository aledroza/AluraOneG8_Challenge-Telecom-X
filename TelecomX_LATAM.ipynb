### Normalizando los dicionarios en Json

```python

df_normalized_customer = pd.json_normalize(df_telecom_X['customer'])
df_normalized_customer
```

```python
df_normalized_phone = pd.json_normalize(df_telecom_X['phone'])
df_normalized_phone
```

```python
df_normalized_internet = pd.json_normalize(df_telecom_X['internet'])
df_normalized_internet
```

```python
df_normalized_account = pd.json_normalize(df_telecom_X['account'])
df_normalized_account
```

#### ESCRIBIENDO JSON NORMALIZADO

```python
df_normalized_customer.to_json('telecom_X_customer_normalizado.json')
df_normalized_phone.to_json('telecom_X_phone_normalizado.json')
df_normalized_internet.to_json('telecom_X_internet_normalizado.json')
df_normalized_account.to_json('telecom_X_account_normalizado.json')
```
