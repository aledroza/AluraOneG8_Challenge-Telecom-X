###**AÃ±adiendo Columnas**

```python
df_telecom_X['gender']= df_normalized_customer['gender']
df_telecom_X
```

```python
df_telecom_X['phone_service']= df_normalized_phone['PhoneService']
df_telecom_X
```

```python
df_telecom_X['contract']= df_normalized_account['Contract']
df_telecom_X
```

```python
df_telecom_X['charges.monthly']= df_normalized_account['Charges.Monthly']
df_telecom_X
```

```python
df_telecom_X['charges.total']= df_normalized_account['Charges.Total']
df_telecom_X
```

```python
df_telecom_X.info()
```

####**Creando Columna cuentas Diarias**

```python
df_telecom_X['cuentas_diarias']= df_normalized_account['Charges.Monthly']/30
df_telecom_X
```

###**Estandarizacion y Transformacion de datos **

```python
df_telecom_X['churn_estandarizado']= df_telecom_X['churn'].str.replace('No','0')
df_telecom_X['churn_estandarizado']= df_telecom_X['churn_estandarizado'].str.replace('Yes','1')
df_telecom_X
```

```python
df_telecom_X['phone_service_estandarizado']=df_telecom_X['phone_service'].apply(lambda x: 0 if x == 'No' else 1)
df_telecom_X
```

```python
df_telecom_X['gender_estandarizado']=df_telecom_X['gender'].apply(lambda x: 0 if x == 'Male' else 1)
df_telecom_X
```

```python
df_telecom_X.drop('churn_estandarizado2', axis=1, inplace=True)
df_telecom_X
```

####**Creando Columna Descripcion**

```python
df_telecom_X['descripcion']= 'El cliente con el ID =' +  df_telecom_X['customerid'] + ',' + 'de genero ='+ '' + df_telecom_X['gender'] + ',' +'con servicio telefonico =' + '' + df_telecom_X['phone_service'] + ',' + 'con contracto por =' + '' + df_telecom_X['contract']  + ',' + 'con cargo mensual de =' + '' + df_telecom_X['charges.monthly'].astype(str) + ',' + 'y un cargo total =' + ',' + df_telecom_X['charges.total'].astype(str) + ',' + 'y cuenta diaria de = ' + ',' + df_telecom_X['cuentas_diarias'].astype(str) + ',' +'tiene evasion / se retiro =' + ',' + df_telecom_X['churn']
df_telecom_X.iloc[500].descripcion
```
