#####**Calculando la Media**

######**La mediana en estadÃ­stica es el valor medio en un conjunto de datos ordenados. Representa el punto medio, donde la mitad de los datos son menores y la otra mitad son mayores.
Â¿CÃ³mo se calcula la mediana?
Ordena los datos: Organiza el conjunto de datos de menor a mayor.
Encuentra el valor medio:
Si el nÃºmero de datos es impar, la mediana es el valor central.
Si el nÃºmero de datos es par, la mediana es el promedio de los dos valores centrales, segÃºn Khan Academy.
Ejemplo:
Para los datos: 2, 3, 3, 5, 7, 10, la mediana es 4 (el promedio de 3 y 5). **
